```python
genero_cargo = dados.groupby(['GENERO', 'CARGO ATUAL'])
genero_cargo
```

```python
def calcular_intervalo(grupo):
    salarios = grupo['SALARIO']
    media_amostral = np.mean(salarios)
    desvio_amostral = np.std(salarios, ddof=1)
    tamanho_amostra = len(salarios)
    nivel_confianca = 0.90
    erro_padrao = stats.sem(salarios)
    intervalo_confianca = stats.t.interval(nivel_confianca, tamanho_amostra - 1, loc=media_amostral, scale=erro_padrao)
    return media_amostral, intervalo_confianca
```

```python
resultados = genero_cargo.apply(lambda grupo: pd.Series(calcular_intervalo(grupo), index=['MEDIA_SALARIAL', 'INTERVALO_CONFIANCA'])).reset_index()
```

```python
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer #quarto passo
```

```python
cramer_coeficiente(dados['cor_raca_etnia'], dados['NIVEL DE ENSINO'])
```
